```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
df = pd.DataFrame(np.random.randint(score(), size=(10,4)), index=[id() for i in range(10)], columns=["국어", "영어", "수학", "사회"])
```

```python
# 과학이 추가되었습니다. 기존 학생들에게 각 랜덤으로 추가하시오.
df.loc[:,'과학'] = pd.Series([score() for i in range(1, 11)], index = df.index)
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럼을 추가하세요.
df.loc[:,'총합'] = pd.Series((df['국어']+df['영어']+df['수학']+df['사회']+df['과학']), index = df.index)
df
```

```python
df.loc['과목합계',:] = pd.Series(df.sum())
df
```

```python
df =df.drop(index=['총합'])
df
```
